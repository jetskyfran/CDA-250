```python
# Print out the basic statistics of the data. For numeric data, this should include
# the mean, median, mode, range, standard deviation, and interquartile (25%, 50%, 75%) range.
# For categorical data, it should include the number of unique values.

print(df.describe())
```

```python
# For each variable, indicate the number of missing values.

print(df.isnull().sum())
```

```python
# Choose two variables. Calculate the skew and kurtosis for each of these variables.

from scipy.stats import skew, kurtosis

revenue_skew = df["revenue"].skew()
revenue_kurt = df["revenue"].kurtosis()

pop_skew = df["popularity"].skew()
pop_kurt = df["popularity"].kurtosis()

print("Revenue Skew:", revenue_skew)
print("Revenue Kurtosis:", revenue_kurt)
print("Popularity Skew:", pop_skew)
print("Popularity Kurtosis:", pop_kurt)
```

```python
# Choose two variables. Represent one variable in a histogram. In the other,
# filter it to represent multiple box plots.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (12, 5))

# histogram:
plt.subplot(1, 2, 1)
sns.histplot(df['popularity'], bins = 30, kde = True, color = 'red')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity')

# boxplot for revenue & popularity:
plt.subplot(1, 2, 2)
sns.boxplot(x = 'genres', y = 'revenue', data = df)
plt.xlabel('Popularity')
plt.ylabel('Runtime')
plt.title('Popularity vs Runtime')

plt.show()
```
